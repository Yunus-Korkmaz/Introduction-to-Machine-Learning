```python
# read data 
import pandas as pd 

data=pd.read_csv("data.csv")
data.head()
```

```python
print(data.info())
```

```python
#look for any NaN value 
data.isna().sum()
```

```python
#drop last column whic we dont need 
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
data.info()
```

```python
data.describe()
```

```python
data.corr()
```

```python
# look for how many malignant and benign tumor
import seaborn as sns 

sns.countplot(x="diagnosis", data=data)
data.loc[:,'diagnosis'].value_counts()
```

```python

x=data.drop("diagnosis",axis=1)
y=data.diagnosis

# veriyi normalize ediyoruz 
from sklearn.preprocessing import Normalizer

norm=Normalizer()
x=norm.fit_transform(x)

```

```python
# setteki diagnosis verilerinin train ve test bölümlerine eşit şekilde bölünmeleri önemli doğru dağılımda setlerdeki diagnosisoranı m=%63 b=%37 şeklinde olmalıdır

# düz bir şekilde verileri kendi içinde karıştırmadan dağıttığımızda diagnosis verileri dengesiz bir şekilde dağılmış oluyor 
# kendi içinde karıştırarak dağıtırdak bu sorunu büyük ölçüde çözebiliriz veya katmanlı dağıtım yapabiliriz 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=False)

print(y_train.value_counts()/len(y_train))
print(y_test.value_counts()/len(y_test))
```

```python
# katmanlı dağıtıtım kullanarak verilerin daha doğru dağılmasını sağlayabiliriz 


norm_data=pd.DataFrame(data=x,columns=[data.columns[data.columns !="diagnosis"]])
norm_data=pd.concat([norm_data,y],axis=1)

from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=42)
for train_index, test_index in split.split(norm_data, norm_data["diagnosis"]):
    
    strat_train_set = norm_data.loc[train_index]
    strat_test_set = norm_data.loc[test_index]

print(strat_test_set["diagnosis"].value_counts() / len(strat_test_set))
print(strat_train_set["diagnosis"].value_counts() / len(strat_train_set))

x_test,y_test=strat_test_set.drop("diagnosis",axis=1).to_numpy(),strat_test_set.diagnosis
x_train,y_train=strat_train_set.drop("diagnosis",axis=1).to_numpy(),strat_train_set.diagnosis
```

```python
x_train
```

```python
# model eğitimi ve accuracy değerinin ölçülmesi 
from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)

print("Accurays is ",knn.score(x_test,y_test))
```

```python
# cross validation ile accuracy değerinin ölçülmesi 
from sklearn.model_selection import cross_val_score
knn=KNeighborsClassifier(n_neighbors=10)

scores=cross_val_score(knn,x,y,cv=3)
print("Scores ",scores)
print("Average score ",scores.mean())
```

```python
# confusion matrix inin sklearn confusion_matrix ile hesaplanması 
from sklearn.metrics import confusion_matrix

knn.fit(x_train,y_train)
x_pred=knn.predict(x_test)

cf_matrix = confusion_matrix(x_pred,y_test)
print(cf_matrix)
```

```python
# confision matrix 
import matplotlib.pyplot as plt
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

ax.set_title('Confusion Matrix\n\n')
ax.set_xlabel('\nPredicted Categorys')
ax.set_ylabel('Actual Categorys ')

plt.show()
```
